```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np

import gmaps
from config import gkey
gmaps.configure(api_key=gkey)
```

```python
filepath='../Project-1/breweries_by_year.csv'
brew_years=pd.read_csv(filepath)
brew_years.head()
```

```python
brew_years.fillna('0',inplace=True)
brew_years.replace({'*':'0'},inplace=True)
brew_years.head()
```

```python
brew_years_reduce=brew_years.drop(['1984','1985','1986','1987','1988','1989','1990','1991','1992',
                                   '1993','1994','1995','1996','1997','1998'],axis=1)
#could also fill values unwanted rows with nan and drop those, Nate
brew_years_reduce.head()
```

```python
filepath2='../Project-1/breweries_us.csv'
us_brew=pd.read_csv(filepath2)
state_brew=us_brew[((us_brew.state=='california')|(us_brew.state=='washington-dc')|(us_brew.state=='tennessee')|
                   (us_brew.state=='new-mexico')|(us_brew.state=='georgia')|(us_brew.state=='new-york'))]
state_brew=state_brew.reset_index(drop=True)
state_brew.head()
```

```python
addy_brew=state_brew.address.str.split(',',expand=True)
addy_brew.columns=['address','city','state','zip','zip2']
#addy_brew=addy_brew.['state'].replace({'DC':'washington-dc'},inplace=True)
#addy_brew=addy_brew.loc[addy_brew['state'].isin(['DC'])]='washington-dc'
addy_brew.head()
```

```python
#mege data frame, tried on state but didn't work, concat on the axis
state_brew2=pd.concat([state_brew,addy_brew],axis=1)
state_brew2.head(5)
```

```python
#cities=['Nashville','Atlanta','New York', 'Albaquerque','San Diego','Washington']
#six_brew=state_brew2.drop(state_brew2[state_brew2['city'] < 25].index, inplace = True) 
#six_cities = state_brew2[~state_brew2['city'].isin(cities)]
#six_cities=state_brew2.loc[(state_brew2['city'] == 'Nashville') & (state_brew2['city'] == 'Atlanta') &
                           #(state_brew2['city'] == 'New York') & (state_brew2['city'] == 'Albaquerque')&
                           #(state_brew2['city'] == 'San Diego') & (state_brew2['city'] == 'Washington')]
six_city=state_brew2[(state_brew2.city==' Nashville')|(state_brew2.city==' Atlanta')|(state_brew2.city==' New York')|
                   (state_brew2.city==' Albuquerque')|(state_brew2.city==' San Diego')|(state_brew2.city==' Washington')]
#six_brew=state_brew2.reset_index(drop=True)
        
six_city.head()
#six_city.city.unique() found out that there was a space before the city
```

```python
six2=six_city.drop(['website','zip2'], axis=1)
six2.head()
```

```python
cols=pd.Series(six2.columns)

for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [dup + '.' + str(i) if i != 0 else dup for i in range(sum(cols == dup))]

# rename the columns with the cols list.
six2.columns=cols
six2.head()
```

```python
six3=six2.drop(['state'], axis=1)
six3.rename(columns={'brewery_name':'Name','type':'Type','address':'Full Address','state_breweries':'Breweries in State (2020)',
                     'address.1':'Street Addresss','city':'City','state.1':'State','zip':'Zip'},inplace=True)

six3.head()
```

```python
six4=six3.groupby('City').count()
six4=six4.drop(['Type','Full Address','Breweries in State (2020)','Street Addresss','State','Zip'], axis=1)
six4.rename(columns={'Name':'Number of Breweries (2020)'},inplace=True)
six4
```

```python
cities13=six_years['Sum of 2013'].astype(float)
cities16=six_years['Sum of 2016'].astype(float)

coords=[(36.174465, -86.767960),(38.893452,-77.014709),(33.748997,-84.387985),
        (35.084385,-106.650421),(40.712776,-74.005974),(32.715736, -117.161087)]
heat_layer=gmaps.heatmap_layer(coords, weights=(cities13),dissipating=False, 
                               max_intensity=100,point_radius = 1)
heat_layer2=gmaps.heatmap_layer(coords, weights=(cities16),dissipating=False, 
                               max_intensity=100,point_radius = 1)
figure_layout={'width':'800px','height':'500px','border':'1px solid black','padding':'1px','margin':'0 auto 0 auto'}

#markers=gmaps.marker_layer('San Deigo','Nashville','Atlanta','Washington','Albuquerque','New York')
#fig.add_layer(markers)
fig=gmaps.figure(layout=figure_layout)
fig.add_layer(heat_layer)
fig.add_layer(heat_layer2)
fig
```
