```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house = pd.read_csv("Input/housing_prices.csv")
```

```python
house.head()
```

```python
#drop unnecessary columns
house_df= house.drop(columns=["RegionType","StateName","Metro","CountyName"])
house_df.head()
```

```python
#create city column from regionname and state
house_df["City"] = house_df["RegionName"] + " " + house_df["State"]
house1_df = house_df.set_index(["City"])
house1_df
```

```python
house2_df = house1_df.drop(columns=["RegionName","State"])
house2_df
```

```python
#isolate  the six cities we chose
cities = ["Nashville TN","New York NY","Atlanta GA","San Diego CA","Albuquerque NM","Washington DC"]

house_set = house2_df.loc[cities]
house_set
```

```python
house_cities = house_set.transpose()

#took out the state in column headings
house_cities2 = house_cities.rename(columns={"Nashville TN": "Nashville","New York NY":"New York","Atlanta GA":"Atlanta","San Diego CA":"San Diego","Albuquerque NM":"Albuquerque"})
house_cities2
```

```python
#dropped some time frames we dont need since just doing 2012-2016
house_cities4 = house_cities2.drop(['11/30/2017','10/31/2017','9/30/2017','8/31/2017','7/31/2017','6/30/2017','5/31/2017','1/31/2017','2/28/2017','3/31/2017','4/30/2017','12/31/2017'])
house_cities4
```

```python
#calc average for 2012 and add it to a new dataframe
cities = ["Nashville","New York","Atlanta","San Diego","Albuquerque","Washington DC"]

df12 = pd.DataFrame(house_cities4.iloc[0:12])
averages12 = pd.DataFrame([df12.mean(axis=0)], columns=list(cities))
averages12
```

```python
#calc average for 2013 and add to the new dataframe
df13 = pd.DataFrame(house_cities4.iloc[12:24])
averages13 = pd.DataFrame([df13.mean(axis=0)], columns=list(cities))
averages = averages12.append(averages13)
averages
```

```python
#same for 2014
df14 = pd.DataFrame(house_cities4.iloc[24:36])
averages14 = pd.DataFrame([df14.mean(axis=0)], columns=list(cities))
averages = averages.append(averages14)
averages
```

```python
#same for 2015
df15 = pd.DataFrame(house_cities4.iloc[36:48])
averages15 = pd.DataFrame([df15.mean(axis=0)], columns=list(cities))
averages = averages.append(averages15)
averages
```

```python
#same for 2016
df16 = pd.DataFrame(house_cities4.iloc[48:60])
averages16 = pd.DataFrame([df16.mean(axis=0)], columns=list(cities))
averages = averages.append(averages16)
averages
```

```python
#add years as index column
years = ['2012','2013','2014','2015','2016']
averages["Years"] = years
averages_df = averages.set_index('Years')
averages_df
```

```python
#plot line graph of housing price by month
house_cities4.plot(kind="line", figsize=(10,7))
plt.xticks(rotation='vertical')
plt.legend(loc='center right')
plt.title("Average Housing Price in Gentrified Areas")
plt.xlabel("Months 2012-2016")
plt.ylabel("Price ($)")
plt.show()
```

```python
lines = averages_df.plot.line(figsize=(6,5))
plt.title("Average Housing Price in Gentrified Areas")
plt.legend(loc='center right')
plt.xlabel("Years 2012-2016")
plt.ylabel("Price ($)")
plt.show()
```

```python
bar = averages_df.plot.bar(figsize=(8,7))
plt.title("Average Housing Price in Gentrified Areas")
plt.legend(loc='upper left')
plt.xlabel("Years 2012-2016")
plt.ylabel("Price ($)")
plt.show()
```
