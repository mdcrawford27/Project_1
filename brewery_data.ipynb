```python
#cities=['Nashville','Atlanta','New York', 'Albaquerque','San Diego','Washington']
#six_brew=state_brew2.drop(state_brew2[state_brew2['city'] < 25].index, inplace = True) 
#six_cities = state_brew2[~state_brew2['city'].isin(cities)]
#six_cities=state_brew2.loc[(state_brew2['city'] == 'Nashville') & (state_brew2['city'] == 'Atlanta') &
                           #(state_brew2['city'] == 'New York') & (state_brew2['city'] == 'Albaquerque')&
                           #(state_brew2['city'] == 'San Diego') & (state_brew2['city'] == 'Washington')]
six_city=state_brew2[(state_brew2.city==' Nashville')|(state_brew2.city==' Atlanta')|(state_brew2.city==' New York')|
                   (state_brew2.city==' Albuquerque')|(state_brew2.city==' San Diego')|(state_brew2.city==' Washington')]
#six_brew=state_brew2.reset_index(drop=True)
        
six_city.head(5)
#six_city.city.unique()
```

```python
six2=six_city.drop(['website','zip2'], axis=1)
six2.head(3)
```

```python
cols=pd.Series(six2.columns)

for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [dup + '.' + str(i) if i != 0 else dup for i in range(sum(cols == dup))]

# rename the columns with the cols list.
six2.columns=cols
six2.head(3)
```

```python
six3=six2.drop(['state'], axis=1)
six3.rename(columns={'brewery_name':'Name','type':'Type','address':'Full Address','state_breweries':'Breweries in State (2020)',
                     'address.1':'Street Addresss','city':'City','state.1':'State','zip':'Zip'},inplace=True)
#Breweries by city
#six3.insert(4, "Breweries in City(2020)",[0], True)

six3
```
